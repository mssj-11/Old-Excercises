# **Proyecto 1: Análisis exploratorio de datos de una encuesta:**

Utilizando una encuesta existente. Para explorar el tema de los alimentos, otras opciones hábitos de ejercicio, opiniones políticas, etc.

Supongamos que has realizado una encuesta sobre las preferencias de comida de tus amigos. Tienes un archivo CSV que contiene los datos de la encuesta con las siguientes columnas:

* Nombre: El nombre del encuestado
* Edad: La edad del encuestado
* Género: El género del encuestado
* Comida Favorita: La comida favorita del encuestado
* Tipo de comida: Si el encuestado prefiere comida vegetariana, vegana, sin gluten, etc.
Para empezar, debes importar las bibliotecas necesarias en Python:

```python
# Librerias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# cargar los datos del archivo CSV en un dataframe de pandas
df = pd.read_csv('datos_encuesta_comida.csv')
```

```python
# explorar los datos con algunos métodos de pandas
# Muestra las primeras 5 filas del dataframe
print(df.head())

# Muestra información sobre las columnas del dataframe
print(df.info())

# Muestra estadísticas descriptivas del dataframe
print(df.describe())
```

## Visualizar los datos con Seaborn

```python
# Visualiza las distribuciones de edad y género
sns.histplot(data=df, x="Edad", hue="Género", multiple="stack")
plt.show()

# Visualiza la distribución de los tipos de comida preferidos
sns.countplot(data=df, x="Tipo de comida")
plt.show()

# Visualiza la distribución de las comidas favoritas
sns.countplot(data=df, x="Comida Favorita")
plt.xticks(rotation=90)
plt.show()
```

## Con esto  se puede ver que la mayoría de los encuestados prefieren comida sin gluten y que la pizza es la comida favorita más popular.
# **Análisis por género**

```python
# Visualiza las distribuciones de edad por género
sns.histplot(data=df, x="Edad", hue="Género", multiple="dodge")
plt.show()

# Visualiza la distribución de los tipos de comida preferidos por género
sns.countplot(data=df, x="Tipo de comida", hue="Género")
plt.show()

# Visualiza la distribución de las comidas favoritas por género
sns.countplot(data=df, x="Comida Favorita", hue="Género")
plt.xticks(rotation=90)
plt.show()
```

### Con estas visualizaciones, se puede ver que los hombres y las mujeres tienen preferencias de comida diferentes y que las mujeres prefieren la comida vegetariana y vegana más que los hombres.
