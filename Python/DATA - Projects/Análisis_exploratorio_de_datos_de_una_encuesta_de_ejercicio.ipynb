# **Análisis exploratorio de datos de una encuesta de ejercicio**

## Ejercicio propuesto:
Supongamos que has realizado una encuesta sobre los hábitos de ejercicio de tus amigos. Tienes un archivo CSV que contiene los datos de la encuesta con las siguientes columnas:
* Nombre: El nombre del encuestado
* Edad: La edad del encuestado
* Género: El género del encuestado
* Tipo de ejercicio: El tipo de ejercicio que prefiere el encuestado (correr, nadar, levantamiento de pesas, etc.)
Frecuencia de ejercicio: Con qué frecuencia hace ejercicio el encuestado (diariamente, semanalmente, mensualmente, etc.)
* Duración del ejercicio: Cuánto tiempo hace ejercicio el encuestado en cada sesión.

```python
# Librerias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# cargar los datos del archivo CSV en un dataframe de pandas
df = pd.read_csv('datos_encuesta_ejercicio.csv')
```

```python
# explorando los datos con algunos métodos de pandas
# Muestra las primeras 5 filas del dataframe
print(df.head())

# Muestra información sobre las columnas del dataframe
print(df.info())

# Muestra estadísticas descriptivas del dataframe
print(df.describe())
```

## Visualizacion los datos con Seaborn:

```python
# Visualiza la distribución de la frecuencia de ejercicio
sns.countplot(data=df, x="Frecuencia de ejercicio")
plt.show()

# Visualiza la distribución de los tipos de ejercicio preferidos
sns.countplot(data=df, x="Tipo de ejercicio")
plt.xticks(rotation=90)
plt.show()

# Visualiza la distribución de la duración del ejercicio
sns.histplot(data=df, x="Duración del ejercicio")
plt.show()
```

## Análisis por género

```python
# Visualiza la distribución de la frecuencia de ejercicio por género
sns.countplot(data=df, x="Frecuencia de ejercicio", hue="Género")
plt.show()

# Visualiza la distribución de los tipos de ejercicio preferidos por género
sns.countplot(data=df, x="Tipo de ejercicio", hue="Género")
plt.xticks(rotation=90)
plt.show()

# Visualiza la distribución de la duración del ejercicio por género
sns.histplot(data=df, x="Duración del ejercicio", hue="Género")
plt.show()
```
