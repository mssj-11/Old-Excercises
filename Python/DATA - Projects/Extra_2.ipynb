# Extracción de información de la pandemia CV19

```python
# Librerias
import pandas as pd
from requests import get
from bs4 import BeautifulSoup
import re
from datetime import datetime
```

```python
# Sitio Web
url_web = 'https://www.worldometers.info/coronavirus/'
```

```python
# Llamando al sitio web
response = get(url_web)

# Convertir los datos a texto
mysoup = BeautifulSoup(response.text, 'html.parser')

# Busqueda de parrafos, <p> es una etiqueta HTML en la que se agregan parrafos
mysoup.find('p')
mysoup.find(id='p')
mysoup.find('div')
# Obtendion de toda la infoamcion guardada en la clase = main_table_countries_div
covid = mysoup.find_all('div', class_='main_table_countries_div')
# Obtención de la tabla
table_cv = mysoup.find('div', id='nav-today')
table_cv = mysoup.find('div', class_='main_table_countries_div')
```

```python
# Llamando a la Tabla
table_cv
```

```python
# Convertir los datos a String
table_cv = str(table_cv)
```

```python
# Llamar a la tabla
table_cv
```

```python
# Convertir los caracteres(<.*?>) a MAYUSCULAS, mediante un recorrido
table_cv = re.sub(r'<.*?>', lambda g: g.group(0).upper(), table_cv)

# Llamar a la tabla
table_cv
```

```python
# Leer los datos y guardarlos en un DataFrame
df = pd.read_html(table_cv)[0]

# Devolviendo el DataFrame
df.head()
```

```python
# Eliminando los valores NAN(nulos) del DataFrame
df = df.dropna()

# Mostrando el DataFrame
df
```
