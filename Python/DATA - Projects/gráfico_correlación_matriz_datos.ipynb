# 1. Importar Librerías:

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

# 2. Generación de Datos Aleatorios:

```python
# Generación de una matriz de datos aleatorios con una correlación conocida
np.random.seed(0)
data = np.random.rand(1000, 5)
data = pd.DataFrame(data, columns=['A', 'B', 'C', 'D', 'E'])
```

* Se genera una matriz de datos aleatorios con 1000 filas y 5 columnas.
* Se convierte en un DataFrame de pandas con columnas etiquetadas como 'A', 'B', 'C', 'D', 'E'.

# 3. Cálculo de la Matriz de Correlación:



```python
correlation_matrix = data.corr()
```

Se calcula la matriz de correlación del DataFrame usando el método `.corr()` de pandas.


# 4. Visualización de la Matriz de Correlación:



```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Matriz de Correlación')
plt.show()
```

* Se crea una figura de tamaño 10x8.
* Se utiliza `seaborn.heatmap` para crear un mapa de calor de la matriz de correlación.
* Se añade el título 'Matriz de Correlación' y se muestra el gráfico.
