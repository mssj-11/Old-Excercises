# **Análisis exploratorio de datos de una encuesta de opiniones políticas**

## Ejercicio propuesto:
Supongamos que has realizado una encuesta sobre opiniones políticas de tus amigos.
Tienes un archivo CSV que contiene los datos de la encuesta con las siguientes columnas:
* Nombre: El nombre del encuestado
* Edad: La edad del encuestado
* Género: El género del encuestado
* Afiliación política
* Opinión sobre el gobierno
* Opinión sobre la economía

## **Matplotlib y Seaborn para graficar histogramas**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar los datos desde el archivo CSV
encuesta = pd.read_csv('datos_encuesta_politica.csv')

# Imprimir las primeras filas de la tabla
print(encuesta.head())

# Describir las variables numéricas
print(encuesta.describe())

# Contar la cantidad de cada afiliación política
print(encuesta['Afiliación política'].value_counts())

# Graficar un histograma de las edades
sns.histplot(encuesta['Edad'], kde=False)
plt.title('Distribución de Edades')
plt.xlabel('Edad')
plt.ylabel('Cantidad')
plt.show()

# Graficar un diagrama de barras de las opiniones sobre el gobierno
sns.countplot(x='Opinión sobre el gobierno', data=encuesta)
plt.title('Opiniones sobre el Gobierno')
plt.xlabel('Opinión')
plt.ylabel('Cantidad')
plt.show()

# Graficar un diagrama de barras apilado de las opiniones sobre el gobierno por afiliación política
sns.countplot(x='Afiliación política', hue='Opinión sobre el gobierno', data=encuesta)
plt.title('Opiniones sobre el Gobierno por Afiliación Política')
plt.xlabel('Afiliación política')
plt.ylabel('Cantidad')
plt.show()

# Graficar un diagrama de caja de las edades por afiliación política
sns.boxplot(x='Afiliación política', y='Edad', data=encuesta)
plt.title('Edades por Afiliación Política')
plt.xlabel('Afiliación política')
plt.ylabel('Edad')
plt.show()
```
